```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/UppEvolucaoMensalDeTitulos.csv', sep= ';', encoding='latin1')
df.columns
```

```python
df_ocorrencias = df.groupby(['upp'])['registro_ocorrencias'].sum().reset_index()
df_ocorrencias = df_ocorrencias.sort_values(by='registro_ocorrencias')
df_ocorrencias
```

```python
df_ocorr_array = np.array(df_ocorrencias['registro_ocorrencias'])
q1 = np.percentile(df_ocorr_array,25)
q2 = np.percentile(df_ocorr_array,50)
q3 = np.percentile(df_ocorr_array,75)
df_ocorr_iqr = q3 - q1
df_ocorr_lim_sup = q3 + (1.5*df_ocorr_iqr)
df_ocorr_lim_inf = q1 - (1.5*df_ocorr_iqr)
```

```python
media_upp = np.mean(df_ocorr_array)
mediana_upp = np.median(df_ocorr_array)
distancia_upp = (media_upp - mediana_upp)/mediana_upp
print(f'Media = {media_upp}\nMediana = {mediana_upp}\nDistancia = {distancia_upp}')
```
