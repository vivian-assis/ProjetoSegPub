```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/DOMensalEstadoDesde1991.csv', sep= ';')
```

```python
df = df.loc[(df['ano']>=2014) & (df['ano'] < 2025)]
```

```python
#df.info()
```

```python
#soma do total por ano
df_recuperacao = df.groupby(['ano'])['recuperacao_veiculos'].sum().reset_index()
df_roubo_veic = df.groupby(['ano'])['roubo_veiculo'].sum().reset_index()
df_furto_veic = df.groupby(['ano'])['furto_veiculos'].sum().reset_index()
#renomeando as variáveis
df_rec = df_recuperacao
df_rb = df_roubo_veic
df_f = df_furto_veic
```

```python
#df_rec = df_recuperacao.loc[(df_recuperacao['ano']>=2014)]
#df_rec = df_recuperacao.loc[(df_recuperacao['ano']>=2014)]
#df_rec = df_recuperacao.loc[(df_recuperacao['ano']>=2014)]
```

```python
#transforma dados de ano em string
df_rec['ano']=df_rec['ano'].astype(str)
```

```python
#MÉDIA, MEDIANA E DISTÂNIA DA RECUPERAÇÃO DE VEÍCULOS
array_rec = np.array(df_rec['recuperacao_veiculos'])
media_rec = np.mean(array_rec)
mediana_rec = np.median(array_rec)
distancia_rec = (media_rec - mediana_rec)/mediana_rec
print(f"MÉDIA = {media_rec:.2f}\nMEDIANA = {mediana_rec:.2f}\nDISTÂNCIA = {distancia_rec:.2f}")
```

```python
#GRÁFICO DO TOTAL DE RECUPERÇÃO DE VEÍCULOS POR ANO
plt.figure(figsize=(10,5))
plt.plot(df_rec['ano'], df_rec['recuperacao_veiculos'], marker='o', linestyle='-')
plt.title('RECUPERAÇÃO DE VEÍCULOS NA ÚLTIMA DÉCADA\nMÉDIA = 23189.73\nMEDIANA = 23663.00\nDISTÂNCIA = -0.02')
plt.xlabel('Ano')
plt.ylabel('Total de Veículos Recuperados por Ano')
#plt.grid(True) #Não é recomendado utilizar Grids para visualizaçõa de dados
plt.tight_layout()
plt.show()
```

```python
#MÉDIA, MEDIANA E DISTÂNIA DO ROUBO DE VEÍCULOS
array_rb = np.array(df_rb['roubo_veiculo'])
media_rb = np.mean(array_rb)
mediana_rb = np.median(array_rb)
distancia = (media_rb - mediana_rb)/mediana_rb
print(f"MÉDIA = {media_rb:.2f}\nMEDIANA = {mediana_rb:.2f}\nDISTÂNCIA = {distancia:.2f}")
```

```python
#GRÁFICO DO TOTAL DE ROUBOS DE VEÍCULOS POR ANO
df_rb['ano']=df_rb['ano'].astype(str)
plt.figure(figsize=(10,5))
plt.plot(df_rb['ano'], df_rb['roubo_veiculo'], marker='o', linestyle='-')
plt.title('ROUBO DE VEÍCULOS NA ÚLTIMA DÉCADA\nMÉDIA = 34520.64\nMEDIANA = 31035.00\nDISTÂNCIA = 0.11')
plt.xlabel('Ano')
plt.ylabel('Total de Veículos Roubados por Ano')
#plt.grid(True) #Não é recomendado utilizar Grids para visualizaçõa de dados
plt.tight_layout()
plt.show()
```

```python
#GRÁFICO DISPERSÃO DE ROUBO DE VEÍCULOS POR ANO
plt.figure(figsize=(10,5))
plt.scatter(df_rb['roubo_veiculo'],df_rec['recuperacao_veiculos'])
plt.title('Roubo x Recuperação nos últimos 10 anos')
plt.xlabel('Roubo de Veículos')
plt.ylabel('Recuperação de Veículos furtados')
plt.show()
```

```python
#MÉDIA, MEDIANA E DISTÂNIA DO FURTO DE VEÍCULOS
array_f = np.array(df_f['furto_veiculos'])
media_f = np.mean(array_f)
mediana_f = np.median(array_f)
distancia_f = (media_f - mediana_f)/mediana_f
print(f"MÉDIA = {media_f:.2f}\nMEDIANA = {mediana_f:.2f}\nDISTÂNCIA = {distancia_f:.2f}")
```

```python
#GRÁFICO DO TOTAL DE FURTOS DE VEÍCULOS POR ANO
df_f['ano']=df_f['ano'].astype(str)
plt.figure(figsize=(10,5))
plt.plot(df_f['ano'], df_f['furto_veiculos'], marker='o', linestyle='-')
plt.title('FURTO DE VEÍCULOS NA ÚLTIMA DÉCADA\nMÉDIA = 16053.09\nMEDIANA = 16577.00\nDISTÂNCIA = -0.03')
plt.xlabel('Ano')
plt.ylabel('Total de Veículos Furtados por Ano')
#plt.grid(True) #Não é recomendado utilizar Grids para visualizaçõa de dados
plt.tight_layout()
plt.show()
```

```python
#GRÁFICO DISPERSÃO DE FURTO DE VEÍCULOS POR ANO
plt.figure(figsize=(10,5))
plt.scatter(df_f['furto_veiculos'],df_rec['recuperacao_veiculos'])
plt.title('Furtos x Recuperação nos últimos 10 anos')
plt.xlabel('Furto de Veículos')
plt.ylabel('Recuperação de Veículos furtados')
plt.show()
```
